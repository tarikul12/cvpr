```python
#resize image
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

#orginal image code
pic =cv2.imread('Tarikul picture.jpg')
pic=cv2.cvtColor(pic,cv2.COLOR_BGR2RGB)
pic.shape
plt.imshow(pic)


```

```python
scale =1
new_width =int(pic.shape[1]*scale)
new_height =int(pic.shape[0]*scale)

#resize  image code
pic_resized=cv2.resize(pic,(new_width,new_height))
pic_resized.shape
plt.imshow(pic_resized)
```

```python
pic =cv2.imread('dog.png')
pic=cv2.cvtColor(pic,cv2.COLOR_BGR2RGB)
pic.shape
plt.imshow(pic)
```

```python
#rotate
pic_r =cv2.rotate(pic,cv2.ROTATE_180)
pic_r1 =cv2.rotate(pic,cv2.ROTATE_90_CLOCKWISE)
pic_r2 =cv2.rotate(pic,cv2.ROTATE_90_COUNTERCLOCKWISE)

plt.imshow(pic_r)
plt.imshow(pic_r1)
plt.imshow(pic_r2)
```
