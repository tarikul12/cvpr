```python
import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow

# Step 1: Upload an image
uploaded = files.upload()

# Step 2: Remove background from the uploaded image
def remove_background(image):
    # Convert image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # Apply a threshold to create a binary mask
    _, mask = cv2.threshold(gray, 240, 255, cv2.THRESH_BINARY)
    # Invert the mask
    mask_inv = cv2.bitwise_not(mask)
    # Create a masked image with the background removed
    masked_image = cv2.bitwise_and(image, image, mask=mask_inv)
    return masked_image

# Load the uploaded image
if len(uploaded) > 0:
    file_name = list(uploaded.keys())[0]
    uploaded_image = cv2.imread(file_name)

    # Step 3: Adjust the background color (for example, changing it to blue)
    blue_bg = np.full_like(uploaded_image, (255, 0, 0), dtype=np.uint8)
    final_image = cv2.add(uploaded_image, blue_bg)

    # Display the original and final images
    cv2_imshow(cv2.cvtColor(uploaded_image, cv2.COLOR_BGR2RGB))
    cv2_imshow(cv2.cvtColor(final_image, cv2.COLOR_BGR2RGB))
else:
    print("No image uploaded.")
```
