```python
#pandas series

import pandas as pd
import numpy as np

ser =pd.Series()
print("Pandas Series:",ser)

data =np.array(['g','e','r','f','c','s','a'])
ser =pd.Series(data)
print("Pandas Series:\n",ser)
```

```python
# DataFrame

import pandas as pd
df =pd.DataFrame() #empty dataframe
print(df)
lst =['g','u','k','j','r','d','f']
df =pd.DataFrame(lst)
print(df)
```

```python
#dataframe from dict of ndarray/lists:
import pandas as pd

lst ={'Name':['Tarikul','Ammu','Nishat','Tania','Ordha','Tanzim'],'Age':[21,48,22,32,27,0.7]}

df =pd.DataFrame(lst)
print(df)
```
